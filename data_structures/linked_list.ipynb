```python
class Node:
	def __init__(self, data, nextnode):
		self.data = data
		self.nextnode = nextnode

class LinkedList:
	def __init__(self):
		self.headnode = None
	
	def add_at_front(self, data):
		self.headnode = Node(data, self.headnode)	  

	def add_at_end(self, data):
		if self.headnode == None:
			self.headnode = Node(data, None)
			return
		curr = self.headnode
		while curr.nextnode != None:
			curr = curr.nextnode
		curr.nextnode = Node(data, None)
		
	def delete_node(self, key):
		if self.headnode == None:
			return
		curr = self.headnode
		prev = None
		while curr != None and curr.data != key:
			prev = curr
			curr = curr.nextnode
		if prev is None:
			self.headnode = curr.nextnode
		elif curr != None:
			prev.nextnode = curr.nextnode
			curr.nextnode = None

	def get_last_node(self):
		if self.headnode == None:
			return
		n = self.headnode
		while n.nextnode != None:
			n = n.nextnode
		print(n.data)

	def is_empty(self):
		if self.headnode == None:
			print("Empty")
		else:
			print("Not Empty")

	def print_list(self):
		n = self.headnode
		while n != None:
			if n.nextnode != None:
				print(n.data, end = " => ")
			else:
				print(n.data, end = "")
			n = n.nextnode
		print()


ll = LinkedList()
ll.is_empty()

print()

ll.add_at_front("B")
ll.add_at_front("A")
ll.add_at_end("C")


ll.is_empty()
ll.get_last_node()
ll.print_list()

ll.delete_node("B")
ll.print_list()

# code and comments by github.com/alandavidgrunberg
```
